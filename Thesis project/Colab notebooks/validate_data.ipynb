```python
def plot_histograms(dataset, columns, rows, cols):
    fig, axs = plt.subplots(rows, cols, figsize=(12, 15), dpi=80)

    for i, feature in enumerate(columns):
        ax = axs[i // cols, i % cols]  # Calculate subplot position
        dataset[columns].hist( ax=ax, facecolor='b')
        ax.set_title(f"{feature} Distribution")

    # Adjust spacing between subplots
    plt.tight_layout()
    plt.show()

plot_histograms(df, df.columns, rows=6, cols=5)
```

```python
#To copy the dataset
standardized_df = df.copy()

# Calculate mean and standard deviation for each column
mean_values = standardized_df.mean()
std_values = standardized_df.std()

# Standardize each column using Z-scores
for col in standardized_df.columns:
    standardized_df[col] = (standardized_df[col] - mean_values[col]) / std_values[col]

# Now showing the standardized data
standardized_df.describe().transpose()
```

```python
import numpy as np

# Assuming your DataFrame is named 'standardized_df'
num_features = len(standardized_df.columns)
rows, cols = 6, 5  # Set rows and columns as needed

fig, axes = plt.subplots(rows, cols, figsize=(16, 18))  # Create subplots

for i, ax in enumerate(axes.flat):
    if i < num_features:
        col = standardized_df.columns[i]
        ax.boxplot(standardized_df[col], vert=True, manage_ticks=True)
        ax.set_title(f'Box Plot for {col}')
        ax.set_ylabel('Inter Quantile Range Values')
        ax.set_xticklabels([col])

        quantiles = np.quantile(standardized_df[col], np.array([0.00, 0.25, 0.50, 0.75, 1.00]))
        ax.hlines(quantiles, [0] * quantiles.size, [1] * quantiles.size,
                  color='b', ls=':', lw=0.5, zorder=0)
        ax.set_xlim(0.5, 1.5)
        ax.set_yticks(quantiles)
    else:
        ax.axis('off')  # Hide extra subplots if there are fewer features

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
